```python
path = r'C:\geopandas\data\data.xlsx'
prov = pd.read_excel(path, sheet_name='Sheet1', header=0)
prov
```

```python
prov_ids = list(prov['PROV_CODE'].unique())
print(prov_ids)
```

```python
prov_df = pd.DataFrame()
for id in prov_ids:
    req_prov = requests.get(f'https://edpapim.disaster.go.th/api/eoc/Shelter?province={id}')
    if req_prov.status_code == 200:

        prov_json = json.loads(req_prov.text)
        df = pd.DataFrame(prov_json)
        prov_df = pd.concat([prov_df, df])
prov_df
```

```python
prov_df.to_csv(r'C:\coop_project_fire_spot\Fire-Spot-Thailand\mylab\Shelter.csv', index=False, encoding='cp874')
```
