```python
def json_to_csv(json_path_folder: str):
    
    list_json_file = os.listdir(json_path_folder)
    for file in list_json_file:
        f = open(os.path.join(json_path_folder, file), mode='r', encoding='utf-8')
        json_data = json.load(f)
        data = pd.DataFrame(json_data['rows'])  # ใช้ข้อมูลเฉพาะที่อยู่ใน tag ที่ชื่อว่า rows
        # output_folder_path = os.path.join(json_path_folder, 'csv_files')
        data.to_csv(os.path.join(r'E:\_GIS\dpm_portal', file.replace('.json', '.csv')), index=False, encoding='cp874')  # เดิม path เป็น .json แต่ export เป็น csv
        f.close()

json_to_csv(r'E:\_GIS\dpm_portal_json')
```

```python
def json_to_csv(json_path: str):
    f = open(json_path, mode='r', encoding='utf-8')
    json_data = json.load(f)
    json_data = json.dumps(json_data)
    data = pd.DataFrame(json_data)  # ใช้ข้อมูลเฉพาะที่อยู่ใน tag ที่ชื่อว่า rows
    # data = pd.DataFrame(json_data['rows'])  # ใช้ข้อมูลเฉพาะที่อยู่ใน tag ที่ชื่อว่า rows
    data.to_csv(json_path.replace('.json', '.csv'), index=False, encoding='cp874')  # เดิม path เป็น .json แต่ export เป็น csv
    f.close()

json_to_csv(r'E:\_BigData\disaster_promptdev\json\1.json')
```
